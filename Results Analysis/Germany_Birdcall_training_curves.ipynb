```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import torch
```

```python
training_resultsTransfer_Cnn6nonenonefreeze0 = pd.read_csv("training_resultsTransfer_Cnn6nonenonefreeze0_mAP=0.649.csv")
training_resultsTransfer_Cnn10nonenonefreeze0 = pd.read_csv("training_resultsTransfer_Cnn10nonenonefreeze0_mAP=0.684.csv")
training_resultsTransfer_Cnn14nonenonefreeze0 = pd.read_csv("training_resultsTransfer_Cnn14nonenonefreeze0_mAP=0.652.csv")

iterations = training_resultsTransfer_Cnn6nonenonefreeze0["iteration"].to_numpy()[1:]
CNN6_loss =  training_resultsTransfer_Cnn6nonenonefreeze0["average loss"].to_numpy()[1:]
CNN10_loss =  training_resultsTransfer_Cnn10nonenonefreeze0["average loss"].to_numpy()[1:]
CNN14_loss =  training_resultsTransfer_Cnn14nonenonefreeze0["average loss"].to_numpy()[1:]
fig = plt.figure(figsize=(6, 4), dpi=150)
plt.xlabel('Iterations')
plt.ylabel('Average loss')
plt.grid(True)
plt.plot(iterations, CNN6_loss, "b", label = "CNN6")
plt.plot(iterations, CNN10_loss, "g", label = "CNN10")
plt.plot(iterations, CNN14_loss, "r", label = "CNN14")
plt.legend()
plt.show()
```

```python
validation_resultsTransfer_Cnn6nonenonefreeze0 = pd.read_csv("validation_resultsTransfer_Cnn6nonenonefreeze0_mAP=0.649.csv")
validation_resultsTransfer_Cnn10nonenonefreeze0 = pd.read_csv("validation_resultsTransfer_Cnn10nonenonefreeze0_mAP=0.684.csv")
validation_resultsTransfer_Cnn14nonenonefreeze0 = pd.read_csv("validation_resultsTransfer_Cnn14nonenonefreeze0_mAP=0.652.csv")

iterations = validation_resultsTransfer_Cnn6nonenonefreeze0["iteration"].to_numpy()
CNN6_mAPs =  validation_resultsTransfer_Cnn6nonenonefreeze0["mAP"].to_numpy()
CNN10_mAPs =  validation_resultsTransfer_Cnn10nonenonefreeze0["mAP"].to_numpy()
CNN14_mAPs =  validation_resultsTransfer_Cnn14nonenonefreeze0["mAP"].to_numpy()
print(np.max(CNN6_mAPs))
print(np.max(CNN10_mAPs))
print(np.max(CNN14_mAPs))

fig = plt.figure(figsize=(6, 4), dpi=150)
plt.xlabel('Iterations')
plt.ylabel('mAP')
plt.ylim([0,1.0])
plt.grid(True)
plt.plot(iterations, CNN6_mAPs, "b", label = "CNN6")
plt.plot(iterations, CNN10_mAPs, "g", label = "CNN10")
plt.plot(iterations, CNN14_mAPs, "r", label = "CNN14")
plt.legend()
plt.show()
```
